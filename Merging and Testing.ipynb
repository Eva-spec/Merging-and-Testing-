```python
import pandas as pd
```

```python
# create 5 lists with the same size
names = ['Eva', 'Kamal', 'John', 'Laura', 'Mathew', 'Heena', 'Katrin', 'Shlok', 'Ravi']
species = ['dog', 'cat', 'velociraptor', 'dog', 'penguin', 'squid', 'cat', 'cat', 'horse']
age = [1, 6, 70, 8, 3, 1, 2, 13, 3]
weight = [10, 5, 15, 7, 4, 1, 3, 2, 380]
color = ['brown', 'black', 'blue', 'black', 'black', 'gray', 'white', 'orange', 'white']
```

```python
# create 3 data frames with the values from the lists
df1 = pd.DataFrame( {'name': names[:3],
                     'species': species[:3],
                     'age': age[:3]})
df2 = pd.DataFrame( {'name': names[3:6],
                     'species': species[3:6],
                     'age': age[3:6]})
df3 = pd.DataFrame( {'name': names[6:],
                     'species': species[6:],
                     'age': age[6:]})
print(df1, '\n')
print(df2, '\n')
print(df3)
```

```python
# .concat to join the dataframes, like a 'union all'
df_list = [df1, df2, df3]
df = pd.concat(df_list)
df
```

```python
# test with mismatching and missing columns
df1 = pd.DataFrame( {'name': names[:3],
                     'specie': species[:3],
                     'age': age[:3]})
df2 = pd.DataFrame( {'name': names[3:6],
                     'species': species[3:6],
                     'age': age[3:6]})
df3 = pd.DataFrame( {'name': names[6:],
                     'age - years': age[6:]})
print(df1, '\n')
print(df2, '\n')
print(df3)
```

```python
# concat with mismatching and missing columns
df_list = [df1, df2, df3]
df = pd.concat(df_list)
df
```
